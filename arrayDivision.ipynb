```python
import sys, io
#sys.stdin = io.StringIO("5 3\n2 4 7 3 5\n")

input = sys.stdin.readline
n, x = map(int, input().split())
array = list(map(int, input().split()))
def check(mid,x):
  parts = 1
  current_sum = 0

  for a in array:
      if current_sum + a <= mid:
          current_sum += a
      else:
          parts += 1
          current_sum = a

  return parts <= x

low = max(array)
high = sum(array)

answer = high

while low <= high:
    mid = (low + high) // 2

    if check(mid,x):
        answer = mid
        high = mid - 1
    else:
        low = mid + 1
print(answer)
```
