```python
import sys, io
sys.stdin = io.StringIO("10\n1 4 7 8 10 3 2 5 6 9\n")

input = sys.stdin.readline
n = map(int, input())
array = list(map(int, input().split()))
array.sort()

def median(arr):
    mid = len(arr) // 2
    if len(arr) % 2 == 0:
        return (arr[mid - 1] + arr[mid]) / 2
    else:
        return arr[mid]

mediana = median(array)

total_sum = 0
for a in array:
    total_sum += abs(a - mediana)
print(int(total_sum))
```
