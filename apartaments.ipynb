```python
import sys, io
sys.stdin = io.StringIO("4 3 5\n60 45 80 60\n30 60 75\n")

input = sys.stdin.readline
n,m,k = map(int, input().split())
array_desired = list(map(int, input().split()))
array_desired.sort()
array_apts = list(map(int, input().split()))
array_apts.sort()
i, j = 0,0
gotIt = 0


while (i < n and j < m ):
  if abs(array_desired[i] - array_apts[j]) <= k:
    gotIt += 1
    i += 1
    j += 1
  elif  array_apts[j] < (array_desired[i] - k):
    j += 1
  else:
    i += 1



print(gotIt)
```
