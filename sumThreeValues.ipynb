```python
import sys
import io

#sys.stdin = io.StringIO("5 8\n6 1 5 3 2\n")

input = sys.stdin.readline

n, x = map(int, input().split())
array = list(map(int, input().split()))
results = []
l = 0
r = n -1

while l < r:
    diff = x - (array[l] + array[r])
    if diff == 0 or diff < 0:
        if array[l] > array[r]:
          l+=1
        else:
          r-=1
    else:
      #tem q limpar o result da outra interação
      results.clear()
      results.append(l+1)
      results.append(r+1)
      l+=1
      r-=1

      if array[l] == diff:
        results.append(l+1)
        break
      elif array[r] == diff:
        results.append(r+1)
        break
if results == [] or len(results) != 3:
  print("IMPOSSIBLE")
else:
  results.sort()
  print(*results) #desempacotar
```
