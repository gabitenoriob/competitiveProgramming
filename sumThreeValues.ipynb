```python
import sys
import io

sys.stdin = io.StringIO("4 8\n2 7 5 1\n")

input = sys.stdin.readline

n, x = map(int, input().split())
array = list(map(int, input().split()))
res = []

vector = [(array[i], i +1) for i in range(n)]
vector.sort()
found = False
for i in range(n - 2):
  l = i+1
  r = n -1
  while l < r:
      if vector[l][0] + vector[r][0] < x - vector[i][0]:
        l += 1
      elif vector[l][0] + vector[r][0] > x - vector[i][0]:
        r-=1
      else:
            res = [vector[i][1], vector[l][1], vector[r][1]]
            res.sort()
            print(*res)
            found = True
            break

  if found:
      break

if not found:
    print("IMPOSSIBLE")
```
