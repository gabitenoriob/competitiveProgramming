```python
import sys, io
#sys.stdin = io.StringIO("3 10\n1 2 3\n")

input = sys.stdin.readline
n,x = map(int, input().split())
array = list(map(int, input().split()))

array.sort()
gondolas = 0
l, r = 0, n-1

while l <= r:
    if array[l] + array[r] <= x:
        gondolas += 1
        l += 1
        r -= 1
    elif array[l] + array[r] > x:
        r -= 1
        gondolas += 1
print(gondolas)

```
