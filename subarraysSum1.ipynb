```python
import sys, io
#sys.stdin = io.StringIO("5 7\n2 4 1 2 7\n")

input = sys.stdin.readline
n, x = map(int, input().split())
array = list(map(int, input().split()))

subs = 0
l = 0
soma_atual = 0

for r in range(n):
    soma_atual += array[r]

    while soma_atual > x:
        soma_atual -= array[l]
        l += 1

    if soma_atual == x:
        subs += 1

print(subs)


```
